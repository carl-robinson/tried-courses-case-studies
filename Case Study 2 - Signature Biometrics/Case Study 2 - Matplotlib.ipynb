```python
%matplotlib inline
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns; sns.set()
```

```python
# read data in
filen = 'complexity24Matrice'
data = pd.read_csv(filen + '.csv', header=None)

# show the data
data.head()
```

```python
# read data in
filen = 'complexity24Matrice_Ac_25sig'
data_ac = pd.read_csv(filen + '.csv', header=None)

# show the data
data_ac.head()
```

```python
# reshape df into a 1d numpy array
v_data = np.reshape(data.values, (np.product(data.values.shape),))
v_data_ac = np.reshape(data_ac.values, (np.product(data_ac.values.shape),))
print(v_data.shape)
print(v_data_ac.shape)
```

```python
# plot scatter of values
fig = plt.figure(figsize=(16, 8))
plt.scatter(v_data, v_data_ac)
plt.title('Coordinate complexity against Acceleration complexity - 24 Gaussians')
plt.xlabel('Coordinate complexity')
plt.ylabel('Acceleration complexity')
plt.tight_layout()
plt.show()
fig.savefig('./IMAGES/coordinate_vs_acceleration_complexity.png')
```
